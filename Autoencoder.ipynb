# Defining an autoencoder

```python
class Autoencoder(Model):
    def __init__(self, latent_dim, shape):
        super(Autoencoder, self).__init__()
        self.latent_dim = latent_dim
        self.shape = shape
        self.encoder = tf.keras.Sequential([
            layers.Flatten(),
            layers.Dense(102,activation='relu'),
            layers.Dense(40,activation='relu'),
            layers.Dense(15,activation='relu'),
            layers.Dense(latent_dim, activation='relu'),
        ])
        self.decoder = tf.keras.Sequential([
            layers.Dense(latent_dim,activation='relu'),
            layers.Dense(15,activation='relu'),
            layers.Dense(40,activation='relu'),
            layers.Dense(102, activation='relu'),
            layers.Reshape(shape)
        ])
    def call(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
shape = (train_data_set_y.shape[1],1)
```

# Bottle-neck

```python
latent_dim = 10
```

# Creating an instance of Autoencoder

```python
autoencoder = Autoencoder(latent_dim, shape)
```

# Compiling autoencoder

```python
autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError())
```

# Training autoencoder

```python
autoencoder.fit(train_data_set_y, train_data_set_y,epochs=10000,shuffle=True,validation_split=0.3)
```

# Encoding the output features

# Linear Network

```python
class Linear_Network(Model):
    def __init__(self, latent_dim, shape):
        super(Linear_Network, self).__init__()
        self.latent_dim = latent_dim
        self.shape = shape

        self.Layers = tf.keras.Sequential([
          layers.Dense(6,activation='relu'),
          layers.Dense(8,activation='relu'),
          layers.Dense(8,activation='relu'),
          layers.Dense(latent_dim, activation='relu'),
        ])
    def call(self, x):
        output = self.Layers(x)
        return output
```

```python
linear_network = Linear_Network(latent_dim, shape)
```

# Compiling

```python
linear_network.compile(optimizer='adam', loss=losses.MeanSquaredError())
```
